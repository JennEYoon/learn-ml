```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# np.linspace, linear space  
# np.linspace(start, stop, [num=50, enpoint=True, retstep=False, dtype=None, axis=0])
x1 = np.linspace(0, 2, 20) 
y1 = np.sin(x1)
print(x1, y1)
```

```python
plt.plot(x1, y1)
```

```python
# np.arange(start, stop, stepsize)  similar to linspace
x2 = np.arange(0, 6, .25)
fig, ax = plt.subplots(1)
ax.plot(x2, np.cos(x2))
```

```python
# OOP method ax.plot()

fig = plt.figure()
ax = plt.axes() 
x3 = np.linspace(0, 6, 30)
ax.plot(x3, np.sin(x3), 'or')
```

```python
# OOP subplots  
fig, ax = plt.subplots(2)
# Can only subscprt ax[0], if there is more than one subplot.  
ax[0].plot(x3, np.sin(x3*2), '-ok')
ax[1].plot(x3, np.cos(x3), '-or')
```

```python
# OOP side-by-side  
fig, ax = plt.subplots(1, 2)
ax[0].plot(x3, np.sin(x3), 'k')
ax[1].plot(x3, np.cos(x3), 'or')
```

```python
g1 = np.geomspace(1, 1000, 50)
log1 = np.logspace(1, 4, 50)
g1, log1
```

```python
# OOP with 2 subplots side by side  
fig, ax = plt.subplots(1, 2)
ax[0].plot(np.sin(g1), g1)
ax[1].plot(log1, g1)
```
