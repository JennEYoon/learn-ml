### Three fundamental data structures, Series, DataFrame, Index  


```python
# Series  
# 1-D array of indexed (ordered) data.  

ser1 = pd.Series(['many', 'an', 'apple', 'and', 'pear', 'are', 'sweet'])
print(ser1)
print(ser1.values)
print(ser1.index)
```

```python
ser2 = np.arange(0, 10, 2)
print(ser2)
ser2 = pd.Series(ser2)
print(ser2)
```

```python
print(ser2.values)
print(ser2.index)
```

```python
print(ser2[-2:])
```
